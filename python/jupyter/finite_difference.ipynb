# Finite Difference Examples

## Radial Schrodinger Equation in 1D, for l=0

```python
# Solve the radial schrodinger equation for l=0, with the central difference approximation to the operator.
from IMPRESPython.finite_difference import radial_schrondinger_eq

n_points = 1000
r_min, r_max = 0.0, 20.0
eigenvalues, eigenvectors = radial_schrondinger_eq(n_points=n_points, r_min=r_min, r_max=r_max)
```

```python
""" Eigenvalues
"""
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

for ev in eigenvalues[0:100]:
    plt.hlines(ev-eigenvalues[0], 0, 10, colors='blue', linestyles='solid')

plt.ylabel('Energy (a.u)')
plt.show()
```

```python
""" Plot the first eigenvectors, noting that as there's no basis
the eigenvectors correspond to the full wave function u(r).
"""
import numpy as np
r = np.linspace(0.0, 20.0, num=1000, endpoint=True)

plt.figure(figsize=(8, 6))
for i in range(0, 1):
    plt.plot(r, eigenvectors[:, i], label=f'State {i} (u(r))')

plt.xlabel('r')
plt.ylabel('u(r)')
plt.title('Radial Eigenfunctions')
plt.legend()
plt.show()
```
